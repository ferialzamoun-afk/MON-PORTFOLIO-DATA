# Visualisation du Portfolio avec subplot_mosaic()

Ce notebook met en avant les projets du portfolio à partir du fichier `cv_extrait_et_amelioration.md` et propose une visualisation synthétique grâce à `matplotlib.pyplot.subplot_mosaic()`. Les extraits du CV et les suggestions d'amélioration sont également valorisés.

```python
# 1. Importer les bibliothèques nécessaires
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

# Pour l'affichage Markdown dans le notebook
from IPython.display import Markdown, display
```

```python
# 2. Charger les données du portfolio
# On suppose que les projets sont listés dans le fichier Markdown sous forme de titres ou de listes
portfolio_md_path = Path('cv_extrait_et_amelioration.md')  # Corrigé: chemin relatif au dossier courant

if not portfolio_md_path.exists():
    raise FileNotFoundError(f"Le fichier {portfolio_md_path} est introuvable dans le dossier courant.")

with open(portfolio_md_path, encoding='utf-8') as f:
    md_lines = f.readlines()

# Extraction naïve des projets : lignes commençant par '- ' ou '### '
projects = []
for line in md_lines:
    if line.startswith('- '):
        projects.append(line[2:].strip())
    elif line.startswith('### '):
        projects.append(line[4:].strip())

# Nettoyage des doublons et vides
projects = [p for p in set(projects) if p]

print(f"Nombre de projets détectés : {len(projects)}")
print("Projets :", projects)
```

```python
# 3. Extraire les informations du cv_extrait
# On extrait un résumé du CV (par exemple, les 10 premières lignes)
portfolio_md_path = Path('cv_extrait_et_amelioration.md')  # Corrigé: chemin relatif au dossier courant

if not portfolio_md_path.exists():
    raise FileNotFoundError(f"Le fichier {portfolio_md_path} est introuvable dans le dossier courant.")

with open(portfolio_md_path, encoding='utf-8') as f:
    cv_excerpt = ''.join([next(f) for _ in range(10)])

# Affichage dans le notebook
print(cv_excerpt)
display(Markdown(cv_excerpt))
```

```python
# 4. Lire et afficher le contenu de amelioration.md
amelioration_path = Path('amelioration.md')
if not amelioration_path.exists():
    # Utiliser le même fichier que le CV si amelioration.md n'existe pas
    amelioration_path = Path('cv_extrait_et_amelioration.md')
    if not amelioration_path.exists():
        raise FileNotFoundError("Aucun fichier amelioration.md ou cv_extrait_et_amelioration.md trouvé dans le dossier courant.")

with open(amelioration_path, encoding='utf-8') as f:
    amelioration_content = ''.join([next(f) for _ in range(10)])

print(amelioration_content)
display(Markdown(amelioration_content))
```

```python
# 5. Créer une visualisation avec subplot_mosaic() selon le nombre de projets
n_projects = len(projects)

# Définir une grille adaptée (ex: 2 lignes si >4 projets)
if n_projects <= 4:
    mosaic = [[f'P{i+1}' for i in range(n_projects)]]
else:
    n_cols = (n_projects + 1) // 2
    mosaic = [ [f'P{i+1}' for i in range(n_cols)], [f'P{i+1+n_cols}' for i in range(n_projects-n_cols)] ]

# Ajouter des cases pour le CV et l'amélioration
mosaic[0].insert(0, 'CV')
mosaic[-1].append('AMELIO')

fig, axes = plt.subplot_mosaic(mosaic, figsize=(4*len(mosaic[0]), 4*len(mosaic)))

# Affichage des titres
for i, proj in enumerate(projects):
    key = f'P{i+1}'
    if key in axes:
        axes[key].set_title(proj)
        axes[key].text(0.5, 0.5, proj, ha='center', va='center', fontsize=12)
        axes[key].axis('off')

# Placeholder pour CV et amélioration
if 'CV' in axes:
    axes['CV'].set_title('Extrait CV')
    axes['CV'].text(0.5, 0.5, 'Voir cellule 3', ha='center', va='center', fontsize=10)
    axes['CV'].axis('off')
if 'AMELIO' in axes:
    axes['AMELIO'].set_title('Améliorations')
    axes['AMELIO'].text(0.5, 0.5, 'Voir cellule 4', ha='center', va='center', fontsize=10)
    axes['AMELIO'].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Histogramme des compétences clés vs nombre de projets
from collections import Counter
import re

# 1. Extraire les compétences clés du CV (fichier chef de projet)
cv_path = Path('cv_chef_de_projet_data.md')
with open(cv_path, encoding='utf-8') as f:
    lines = f.readlines()

competences_cles = []
inside_comp = False
for line in lines:
    if line.strip().lower().startswith('## compétences clés'):
        inside_comp = True
        continue
    if inside_comp:
        if line.strip().startswith('##') and not line.strip().lower().startswith('## compétences clés'):
            break
        if line.strip().startswith('- '):
            competences_cles.append(line.strip()[2:])

# 2. Extraire les projets et compétences associées du portfolio (projets_data.md)
projets_path = Path('projets_data.md')
with open(projets_path, encoding='utf-8') as f:
    projets_lines = f.readlines()

projets_competences = []
current_proj = None
for line in projets_lines:
    if line.strip().startswith('### '):
        current_proj = line.strip()[4:]
    if '**Compétences :**' in line:
        comp_line = line.split('**Compétences :**')[-1].strip()
        comp_list = [c.strip().lower() for c in re.split(',|;', comp_line)]
        projets_competences.append((current_proj, comp_list))

# 3. Compter le nombre de projets par compétence clé (recherche partielle insensible à la casse)
comp_counter = Counter()
for comp_cle in competences_cles:
    comp_cle_lower = comp_cle.lower()
    for proj, comp_list in projets_competences:
        if any(comp_cle_lower.split()[0] in c for c in comp_list):
            comp_counter[comp_cle] += 1

# 4. Affichage de l'histogramme dans un subplot
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(comp_counter.keys(), comp_counter.values(), color='skyblue')
ax.set_title("Nombre de projets par compétence clé")
ax.set_ylabel("Nombre de projets")
ax.set_xticklabels(comp_counter.keys(), rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
