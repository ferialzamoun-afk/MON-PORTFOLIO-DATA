- Vérifier la cohérence des données de clients avec les transactions (ex. pas de transactions pour des clients inexistants).

```python
# Clients sans transaction
clients_avec_achat = transactions['client_id'].unique()
clients_sans_achat = customers[~customers['client_id'].isin(clients_avec_achat)]
print(f"\nClients sans transaction : {len(clients_sans_achat)} sur {len(customers)}")
print(clients_sans_achat)
```

```python
# Ajouter une colonne statut_actif dans customers (version simple)
if 'customers' not in globals() or 'transactions' not in globals():
    print("Veuillez d'abord exécuter les cellules de chargement des fichiers (customers + transactions).")
else:
    clients_avec_achat = transactions['client_id'].dropna().unique()
    customers['statut_actif'] = customers['client_id'].isin(clients_avec_achat)
    customers['statut_actif'] = customers['statut_actif'].map({True: 'Actif', False: 'Inactif'})

    print(customers['statut_actif'].value_counts())
    display(customers[['client_id', 'statut_actif']].head(10))
```

```python
# Segmenter les clients en BtoB / BtoC avec un seuil de CA
seuil_ca_btob = 100000

# Ici, transactions ne contient pas encore 'price' dans cette section du notebook.
# On recalcule donc un CA client via une jointure temporaire avec products.
transactions_ca = transactions.merge(
    products[['id_prod', 'price']],
    on='id_prod',
    how='left'
 )

# Sécurité: si des prix sont manquants après jointure, on les met à 0
transactions_ca['price'] = transactions_ca['price'].fillna(0)

# CA total par client
ca_client = transactions_ca.groupby('client_id', as_index=False)['price'].sum()
ca_client.columns = ['client_id', 'ca_total']

# Segment selon le seuil
ca_client['segment_client'] = np.where(
    ca_client['ca_total'] > seuil_ca_btob,
    'BtoB',
    'BtoC'
 )

# Ajouter le segment dans customers
customers = customers.merge(ca_client[['client_id', 'ca_total', 'segment_client']], on='client_id', how='left')
customers['segment_client'] = customers['segment_client'].fillna('Sans achat')
customers['ca_total'] = customers['ca_total'].fillna(0)

print(f"Seuil BtoB : {seuil_ca_btob:,.0f} €")
print("\nRépartition des segments :")
print(customers['segment_client'].value_counts())

print("\nAperçu clients BtoB :")
display(customers[customers['segment_client'] == 'BtoB'][['client_id', 'ca_total', 'segment_client']].head(10))
```

<h3 style="margin: auto; padding: 20px; color: #F4F6F7; ">1.7 - Jointure des tables</h3>
</div>

Les 3 fichiers de données possèdent des clés communes qui permettent de les relier entre eux en respectant les règles d'intégrité référentielle. A ce stade, nous souhaitons conserver toutes les lignes de la table des transactions et ne conserver que les lignes correspondantes dans les tables des produits et des clients. Nous allons donc effectuer des jointures de type "Left join" en utilisant les clés suivantes : client_id pour relier transactions et customers, et id_prod pour relier transactions et products.

```python
# Jointure transactions + products
df = transactions.merge(products, on='id_prod', how='left')


print("Shape après jointure :", df.shape)
print("\nAperçu :\n", df.head())
print("\nNaN après jointure :\n", df.isna().sum())
```

```python
# Jointure avec customers
df = df.merge(customers, on='client_id', how='left')

print("Shape après jointure customers :", df.shape)
print("\nAperçu :\n", df.head())
print("\nNaN après jointure :\n", df.isna().sum())
```

<div style="background-color: #2E5EAA; border: 1px solid #2E5EAA;" >
<h3 style="margin: auto; padding: 20px; color: #F4F6F7; ">1.8 - Feature Engineering</h3>
</div>

- Calcul de l'âge des clients à partir de leur date de naissance pour faciliter les analyses démographiques 
- calcul des données Mois , Année à partir de la date de transaction pour faciliter les analyses temporelles et saisonnières.
- Création d'un libellé de catégorie pour faciliter les analyses de segmentation produits

```python
# Calcul de l'âge
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['birth'] = pd.to_numeric(df['birth'], errors='coerce')
df['age'] = df['date'].dt.year - df['birth']

print("Shape après jointure :", df.shape)
print("\nAperçu :\n", df.head())
print("\nNaN après jointure :\n", df.isna().sum())
```

```python
# Extraction mois et année
df['mois'] = df['date'].dt.to_period('M')
df['annee'] = df['date'].dt.year

print("Aperçu des nouvelles colonnes :\n", df[['date', 'birth', 'age', 'mois', 'annee']].head())
```

```python
# Labels lisibles pour les catégories
categ_labels = {0: 'Catégorie 0', 1: 'Catégorie 1', 2: 'Catégorie 2'}
df['categ_label'] = df['categ'].map(categ_labels)
print("Aperçu des nouvelles colonnes :\n", df[['date', 'birth', 'age', 'mois', 'annee','categ_label']].head())
```
