```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, Circle

# Si Century Gothic est installée sur ta machine, tu peux l'activer ainsi :
from matplotlib import font_manager, rcParams

# Essaie de forcer Century Gothic si dispo (sinon fallback par défaut)
for f in font_manager.findSystemFonts():
    if "GOTHIC" in f.upper() or "CENTURY GOTHIC" in f.upper():
        rcParams['font.family'] = font_manager.FontProperties(fname=f).get_name()
        break

# --- FIGURE ---
plt.figure(figsize=(16, 9))
plt.axis("off")
ax = plt.gca()
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

def draw_block(x, y, w, h, text, color, fontsize=16):
    box = FancyBboxPatch(
        (x, y), w, h,
        boxstyle="round,pad=0.02,rounding_size=0.03",
        linewidth=2,
        edgecolor=color,
        facecolor="white"
    )
    ax.add_patch(box)
    ax.text(
        x + w/2, y + h/2, text,
        ha="center", va="center",
        fontsize=fontsize, color="black"
    )

# --- BLOC CENTRAL ---
center_w, center_h = 0.34, 0.24
center_x, center_y = 0.33, 0.38

center_color = "#1E3A8A"
center_box = FancyBboxPatch(
    (center_x, center_y), center_w, center_h,
    boxstyle="round,pad=0.03,rounding_size=0.04",
    linewidth=2,
    edgecolor=center_color,
    facecolor=center_color
)
ax.add_patch(center_box)

center_text = (
    "Fériel Zamoun\n"
    "Data Analyst – RNCP 37837\n"
    "SQL • Python • Power BI • Statistiques"
)

ax.text(
    center_x + center_w/2, center_y + center_h/2,
    center_text,
    ha="center", va="center",
    fontsize=18, color="white"
)

# --- BLOCS AUTOUR ---

block_w, block_h = 0.30, 0.22

# Haut gauche
draw_block(
    0.05, 0.65, block_w, block_h,
    "🧠 Compétences techniques\n\n"
    "SQL\nPython\nStatistiques\nMachine Learning\nData cleaning",
    "#3B82F6", fontsize=15
)

# Haut droit
draw_block(
    0.65, 0.65, block_w, block_h,
    "📊 Data Visualisation & BI\n\n"
    "Power BI\nDashboards KPI\nReporting\nStorytelling",
    "#10B981", fontsize=15
)

# Bas gauche
draw_block(
    0.05, 0.10, block_w, block_h,
    "📈 Analyse & Modélisation\n\n"
    "EDA\nSegmentation\nModèles prédictifs\nOptimisation KPI",
    "#8B5CF6", fontsize=15
)

# Bas droit
draw_block(
    0.65, 0.10, block_w, block_h,
    "🧭 Gestion de projet & conformité\n\n"
    "Agile\nProjets Data/IA\nDocumentation\nRGPD",
    "#F59E0B", fontsize=15
)

# --- CONNECTEURS (lignes simples) ---
line_color = "#C7C7C7"
line_width = 2

# Centre -> haut gauche
ax.plot(
    [center_x + center_w/2, 0.05 + block_w/2],
    [center_y + center_h, 0.65],
    color=line_color, linewidth=line_width
)

# Centre -> haut droit
ax.plot(
    [center_x + center_w/2, 0.65 + block_w/2],
    [center_y + center_h, 0.65],
    color=line_color, linewidth=line_width
)

# Centre -> bas gauche
ax.plot(
    [center_x + center_w/2, 0.05 + block_w/2],
    [center_y, 0.10 + block_h],
    color=line_color, linewidth=line_width
)

# Centre -> bas droit
ax.plot(
    [center_x + center_w/2, 0.65 + block_w/2],
    [center_y, 0.10 + block_h],
    color=line_color, linewidth=line_width
)

# --- PHRASE SIGNATURE ---
ax.text(
    0.5, 0.04,
    "En reconversion – 15 ans d’expérience en pilotage & qualité • Stage Data Avril 2026",
    ha="center", va="center",
    fontsize=14, color="#444444"
)

# --- SAUVEGARDE ---
plt.savefig("carte_mentale_ferial_century_gothic.png", dpi=300, bbox_inches="tight")
plt.show()
```
